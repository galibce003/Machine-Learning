```python
df_train.drop(['PassengerId','Name','Ticket', 'Cabin'], axis = 1, inplace = True)
```

## Dealing missing values

```python
df_train.isnull().sum()
```

```python
df1_train = df_train.fillna(df_train.mean())
```

```python
df1_train.isnull().sum()
```

```python
df1_train['Embarked'].value_counts()
```

```python
df1_train['Embarked'].fillna(value = 'S', inplace = True)
```

```python
df1_train.isnull().sum()
```

```python
df_test.drop(['Name','Ticket', 'Cabin'], axis = 1, inplace = True)
```

```python
df_test.isnull().sum()
```

```python
df1_test = df_test.fillna(df_test.mean())
```

```python
df1_test['Embarked'].fillna(value = 'S', inplace = True)
```

```python
df1_test.isnull().sum()
```

<br/>

## Converted to dummy

```python
df2_train = pd.get_dummies(df1_train, columns = ['Pclass','Sex','Embarked'], drop_first = True)
df2_test = pd.get_dummies(df1_test, columns = ['Pclass','Sex','Embarked'], drop_first = True)
```

<br/>

## Split into X, y and X_test

```python
X = pd.DataFrame(df2_train[['Age', 'SibSp', 'Parch', 'Fare', 'Pclass_2', 'Pclass_3',
       'Sex_male', 'Embarked_Q', 'Embarked_S']])
y = df2_train[['Survived']]
X_test = pd.DataFrame(df2_test[['Age', 'SibSp', 'Parch', 'Fare', 'Pclass_2', 'Pclass_3',
       'Sex_male', 'Embarked_Q', 'Embarked_S']])
```

<br/>

## Model build up and Prediction

```python
s = SVC(kernel = 'linear', gamma = 'auto', C = 2)
s.fit(X, y)
y_pred = s.predict(X_test)
```

<br/>

## SWave test sample with predicted value

```python
df2_test['Survived'] = list(y_pred)
```
