## Import libraries

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Getting data

```python
x = [2,4,6,8,10,12]
y1 = [5,6,9,12,20,23]
y2 = [8,10,15,21,16,19]
mean_y1 = 12.5
```

<br/>
<br/>

## Fill between two lines

```python
plt.plot(x, y1, label = "y1", color = 'Red')
plt.plot(x, y2, label = "y2", color = 'black')

plt.fill_between(x, y1, y2, color = 'yellow', alpha = 0.55)

plt.legend()
plt.show()
```

<br/>
<br/>

# Fill between a line and bottom

```python
plt.plot(x, y1, label = "y1", color = 'Red', linewidth = 5)
plt.plot(x, y2, label = "y2", color = 'black')

plt.fill_between(x, y1, color = 'yellow', alpha = 0.55)

plt.legend()
plt.show()
```

<br/>
<br/>

## Fill between a line and its mean

```python
plt.plot(x, y1, label = "y1", color = 'Red', linewidth = 5)
plt.plot(x, y2, label = "y2", color = 'black')

plt.fill_between(x, y1, mean_y1, color = 'blue', alpha = 0.49)

plt.legend()
plt.show()
```

<br/>
<br/>

## Fill up by two different colors between 'y1 > mean_y1' and 'y1 <= mean_y1'
#### (y1 should be dataframe to compare with mean_y1. But here y1 is a list.)

```python
plt.plot(x, y1, label = "y1", color = 'Red')
plt.plot(x, y2, label = "y2", color = 'black')

plt.fill_between(x, y1, mean_y1, where = (y1 > mean_y1), color = 'blue', alpha = 0.49, label = "Above mean")
plt.fill_between(x, y1, mean_y1, where = (y1 <= mean_y1), color = 'blue', alpha = 0.49, label = "Below mean")

plt.legend()
plt.show()
```
