## Import libraries

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

<br/>
<br/>

## Data for plotting

```python
slices = [60, 40, 30, 10]
labels = ['Python', 'C', 'Ruby', 'HTML']
color = ['blue', 'red', 'yellow', 'green']
explode = [0.1,0,0,0]
```

<br/>
<br/>

## Simple Pie Plot

```python
plt.pie(slices, labels = labels)
plt.show()
```

<br/>
<br/>

## Adding Edge Color

```python
plt.pie(slices, labels = labels, wedgeprops = {'edgecolor' : 'black'})
plt.show()
```

<br/>
<br/>

## Customize the color

```python
plt.pie(slices, labels = labels, wedgeprops = {'edgecolor' : 'black'}, colors = color)
plt.show()
```

<br/>
<br/>

## Highlight one slice with Shadow

```python
plt.pie(slices, labels = labels, explode = explode, shadow = True)
plt.show()
```

<br/>
<br/>

## Rotate the chart

```python
plt.pie(slices, labels = labels, explode = explode, startangle = 90)
plt.show()
```

<br/>
<br/>

## Add percentage

```python
plt.pie(slices, labels = labels, autopct = '%1.1f%%')
plt.show()
```
