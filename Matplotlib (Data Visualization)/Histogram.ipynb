## Import libraries and getting datas

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ages = [10,15,21,25,26,29,32,38,45,55]
mean_age = 29.6
```

<br/>

## Plot a primary Histogram

```python
plt.hist(ages, edgecolor = 'black')
plt.show()
```

<br/>

## Primary Histogram with specific number of bins

```python
plt.hist(ages, bins = 5, edgecolor = 'black')
plt.show()
```

<br/>

## Primary Histogram with specific bins

```python
bins = [10,20,30,40,50,60]
plt.hist(ages, bins = bins, edgecolor = 'black')
plt.show()
```

<br/>

## Removing a range from 0-10 from plot

```python
bins20 = [20,30,40,50,60]
plt.hist(ages, bins = bins20, edgecolor = 'black')
plt.show()
```

<br/>

## Logarithmic scale

```python
bins20 = [20,30,40,50,60]
plt.hist(ages, bins = bins20, edgecolor = 'black', log = True)
plt.show()
```

<br/>

## Histogram with a vertical line

```python
bins20 = [20,30,40,50,60]
plt.hist(ages, bins = bins20, edgecolor = 'black')

plt.axvline(mean_age, color = 'red', label = 'Mean Age', linewidth = 3)
plt.legend()

plt.show()
```
