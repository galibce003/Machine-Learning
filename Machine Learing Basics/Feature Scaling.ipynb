# Normalization (min-max scalar)
### Normalization scales down to bound all the values between 0 - 1

```python
scaling = MinMaxScaler()
```

```python
scaling.fit_transform(df[['Cement', 'Slag', 'Fly_ash', 'Water', 'SP', 'Coarse_Aggr', 'Fine_Aggr', 'SLUMP(cm)',
                          'FLOW(cm)', 'com_strength']])
```

# Standarization (Z-score Normalization)
### Standarization scales down as mean = 0 and Standard deviation = 1

```python
scaling1 = StandardScaler()
```

```python
scaling1.fit_transform(df[['Cement', 'Slag', 'Fly_ash', 'Water', 'SP', 'Coarse_Aggr', 'Fine_Aggr', 'SLUMP(cm)',
                          'FLOW(cm)', 'com_strength']])
```
