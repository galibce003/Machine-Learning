## Load the dataset and libraries as well

```python
import pandas as pd

df = pd.read_csv('C:/Users/Mehedi Hassan Galib/Desktop/Python/datas/iris.csv')
df.head()
```

<br/>
<br/>

## Convert a column into index

```python
df1 = df.set_index("Id")
df1.head()
```

<br/>
<br/>

## Then Convert the index into column again

```python
df2 = df1.reset_index()
df2.head()
```

<br/>
<br/>

## Multi-level index

```python
df3 = df2.set_index(["Id",'species'])
df3.tail()
```

<br/>
<br/>

## Finding values by indexing

```python
df3.loc[[2,'Iris-setosa']]
```

```python
df3.loc[[(2,'Iris-setosa'), (149,'Iris-virginica')]]
```

<br/>
<br/>

## Sorting Index

```python
df4 = df3.reset_index()
df4.head()
```

```python
df5 = df4.set_index('species')
df5.tail()
```

```python
df5.sort_index()
df5.tail()
```
